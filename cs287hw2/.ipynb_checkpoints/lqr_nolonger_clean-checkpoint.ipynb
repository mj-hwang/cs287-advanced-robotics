Recall the following optimal control problem:
$$\min_{x,u} \sum_{t=1}^T (x_t' Q x_t + u_t' R u_t) + x_{T+1}' Q_{final} x_{T+1} \\ \text{s.t. }x_{t+1} = A x_t + B u_t$$
 

For $T$ going to infinity, we'll have that the Value Function and the
Feedback Controller at time $t=0$ reach a steady-state --- the optimal
value function and feedback controller for infinitely many time-steps to-go

So let's run the Value Iteration Solution to the LQR control problem
until it has converged, and then use that infinite horizon optimal
feedback controller to stabilize our system at 0. 

```python
# implement the infinite horizon optimal feedback controller
def lqr_infinite_horizon(A, B, Q, R):
    """
    find the infinite horizon K and P through running LQR back-ups
    until norm(K_new - K_curr, 2) <= 1e-4
    return: K, P
    """
    
    """YOUR CODE HERE"""
    ds, du = A.shape[0], B.shape[1]
    P, K_current = np.eye(ds), np.zeros((du, ds))
    while True:
        K_new = -np.linalg.pinv(R + B.T @ P @ B) @ B.T @ P @ A
        if np.linalg.norm(K_new - K_current) <= 1e-4:
            break
        K_current = K_new
        A_BK = A + B @ K_new
        P = Q + K_new.T @ R @ K_new + A_BK.T @ P @ A_BK
    return K_new, P
    """YOUR CODE ENDS HERE"""
```

```python
# problem has been defined, let's solve it:
Q, R = np.eye(nX), np.eye(nU)
K_inf, P_inf = lqr_infinite_horizon(A, B, Q, R)
```

#### (b) LQR-based Stabilization for Nonlinear Systems

Now let's consider nonlinear systems. Linearize around one point and design an infinite horizon controller for the resulting system. 

```python
# implement linearization about a point
def linearize_dynamics(f, x_ref, u_ref, dt, my_eps, x_ref_tplus1=None):
    """
    x(:,t+1) - x_ref  approximately = A*( x(:,t)-x_ref ) + B* ( u(:,t) - u_ref ) + c
    if we pick x_ref and u_ref to constitute a fixed point, then c == 0
    return: A, B, c
    """
    
    """YOUR CODE HERE"""
    if x_ref_tplus1 is not None:
        x_ref_next = x_ref_tplus1
    else:
        x_ref_next = x_ref
    ds = x_ref.shape[0]
    A = np.zeros((ds, ds))
    for i in range(ds):
        x = copy.deepcopy(x_ref).astype(float)
        x[i] = x_ref[i] + my_eps
        d1 = f(x, u_ref, dt)
        x = copy.deepcopy(x_ref)
        x[i] = x_ref[i] - my_eps
        d2 = f(x, u_ref, dt)
        A[:,i] = (d1-d2)/(2*my_eps)
    du = u_ref.shape[0]
    B = np.zeros((ds, du))
    for i in range(du):
        u = copy.deepcopy(u_ref).astype(float)
        u[i] = u_ref[i] + my_eps
        d1 = f(x_ref_next, u, dt)
        u = copy.deepcopy(u_ref).astype(float)
        u[i] = u_ref[i] - my_eps
        d2 = f(x_ref_next, u, dt)
        B[:,i] = (d1-d2)/(2*my_eps)
    c = f(x_ref, u_ref, dt) - x_ref_next
    if len(B.shape) == 1:
        return A, B.reshape(-1, 1), c
    
    """YOUR CODE ENDS HERE"""
    return A, B, c
```

```python
# take an environment and find the infinite horizon controller for the linearized system
def lqr_nonlinear(config):
    f = config['env']
    dt = 0.008 # we work with discrete time
    my_eps = 0.008 # finite difference for numerical differentiation
    
    # load in our reference points 
    x_ref, u_ref = config['x_ref'], config['u_ref']
    
    # linearize
    A, B, c = linearize_dynamics(f, x_ref, u_ref, dt, my_eps)
    nX, nU = A.shape[0], B.shape[1]
    Q, R = np.eye(nX), np.eye(nU)
    
    # solve for the linearized system
    K_inf, P_inf = lqr_infinite_horizon(A, B, Q, R) # you implemented in part (a)
    
    # recognize the simulation code from part (a)? modify it to use your controller at each timestep
    def simulate(K_inf, f, x_ref, u_ref, dt, n_starting_states, T, noise=None):
        for s in np.arange(n_starting_states):
            x, u = np.zeros((K_inf.shape[1], T+1)), np.zeros((K_inf.shape[0], T+1))
            x[:,0] = starting_states[:,s]
            for t in np.arange(T):
                """YOUR CODE HERE"""
                u[:,t] = (K_inf @ (x[:,t] - x_ref)) + u_ref
                """YOUR CODE ENDS HERE"""
                x[:,t+1] = f(x[:,t], u[:,t], dt)
                if noise is not None:
                    x[:,t+1] += noise[:,t]
            plt.plot(x.T, linewidth=.6)
            plt.plot(np.squeeze(u), linewidth=.7, linestyle='--')
            plt.show()
        
    # now let's simulate and see what happens for a few different starting states
    starting_states = config['ss']
    n_starting_states = starting_states.shape[1]
    T = config['steps'] # simulating for T steps
    simulate(K_inf, f, x_ref, u_ref, dt, n_starting_states, T)

    # and now in the presence of noise
    noise_id = config['noise']
    simulate(K_inf, f, x_ref, u_ref, dt, n_starting_states, T, noise=loadmat(noise_id+".mat")[noise_id])
```

#### (c) Linear Time Varying (LTV) Systems

Now we'll consider LTV systems, i.e., $x_{t+1} = A_tx_t + B_tu_t$ (and $Q, R$ may be time-dependent). Our optimal control problem is thus:

$$\min_{x,u} \sum_{t=1}^T (x_t' Q_t x_t + u_t' R_t u_t) + x_{T+1}' Q_{T} x_{T+1} \\ \text{s.t. }x_{t+1} = A_t x_t + B_t u_t$$


```python
# implement a finite horizon optimal feedback controller, accounting for possibly time-varying parameters
def lqr_finite_horizon(A_lst, B_lst, Q_lst, R_lst, T):
    """
    find the finite horizon K and P through running LQR back-ups
    return: K_{1:T}, P_{1:T}
    """
    """YOUR CODE HERE"""
    A = (lambda x: A_lst[x]) if type(A_lst) == list else (lambda x: A_lst)
    B = (lambda x: B_lst[x]) if type(B_lst) == list else (lambda x: B_lst)
    Q = (lambda x: Q_lst[x]) if type(Q_lst) == list else (lambda x: Q_lst)
    R = (lambda x: R_lst[x]) if type(R_lst) == list else (lambda x: R_lst)
        
    K_lst, P_lst= [], []
    P, i = Q(0), 0
    for i in range(T):
        H_i = T-i-1
        K = -np.linalg.pinv(R(H_i) + B(H_i).T @ P @ B(H_i)) @ B(H_i).T @ P @ A(H_i)
        A_BK = A(H_i) + B(H_i) @ K
        P = Q(H_i) + K.T @ R(H_i) @ K + A_BK.T @ P @ A_BK
        
        P_lst.append(P)
        K_lst.append(K)
    
    K_lst.reverse()
    P_lst.reverse()
    return K_lst, P_lst
    """YOUR CODE ENDS HERE"""
```

```python
# here we define a LTV system for a fixed horizon
T = 100
A = [np.array([[np.sin(t), -0.5049, 0.0299, 2.6544, 1.0608],
              [2.3846, -0.2312, -0.1260, -0.7945, 0.5279],
              [1.4019, -0.6394, -0.1401, 0.5484, 0.1624],
              [-0.0254, 0.4595, -0.0862, 2.1750, 1.1012],
              [0.5172, 0.5060, 1.6579, -0.9407, -1.4441]]) for t in range(T)]
B = [np.array([[-0.7789, -1.2076],
              [0.4299, -1.6041],
              [0.2006, -1.7395],
              [0.8302, 0.2295],
              [-1.8465, np.cos(t)]]) for t in range(T)]
nX, nU = A[0].shape[0], B[0].shape[1]
```

```python
starting_states = np.array([[-1.9613, 1.9277, -0.2442],
                            [-1.3127, -0.2406, -0.0260],
                            [0.0698, -0.5860, -0.7522],
                            [0.0935, -0.1524, -0.9680],
                            [1.2494, 0.5397, -0.5146]])
n_starting_states = starting_states.shape[1]

Q, R = np.eye(nX), np.eye(nU)
K_lst, P_lst = lqr_finite_horizon(A, B, Q, R, T)

# fill in to use your controller
def simulate(A, B, K_list, n_starting_states, T, noise=None):
    for s in np.arange(n_starting_states):
        x, u = np.zeros((K_list[0].shape[1], T+1)), np.zeros((K_list[0].shape[0], T+1))
        x[:,0] = starting_states[:,s]
        for t in np.arange(T):
            """YOUR CODE HERE"""
            u[:,t] = K_list[t] @ x[:,t]
            """YOUR CODE ENDS HERE"""
            x[:,t+1] = A[t] @ x[:,t] + B[t] @ u[:,t]
            if noise is not None:
                x[:,t+1] += noise[:,t]
        plt.plot(x.T, linewidth=.7)
        plt.show()

# simulate to sanity check your TV solution
simulate(A, B, K_lst, n_starting_states, T)
```

```python
from hopper_env import HopperModEnv
env = HopperModEnv()

def hopper_sim(x0, u, dt, rollout=False, perturb=.01):
    nq, nv = env.model.nq, env.model.nv
    env.sim.reset()
    qpos = copy.deepcopy(env.init_qpos)
    qvel = copy.deepcopy(env.init_qvel)
    
    qpos[1:6] = x0[:nq-1]
    qvel[:6] = x0[nq-1:]
  
    env.set_state(qpos, qvel)
    if rollout:
        env.step(u, perturb=perturb)
    else:
        env.perturb_joints = False
        env.step(u)
        env.perturb_joints = True
    return np.concatenate([
        env.sim.data.qpos.flat[1:6],
        env.sim.data.qvel.flat[:6]
    ])

x_ref, u_ref = np.concatenate([env.init_qpos[1:],env.init_qvel[:]]), np.zeros(env.action_space.sample().shape[0])

hopper_config = {
    'env': env,
    'steps': (500//3)*3,
    'x_ref': x_ref,
    'u_ref': u_ref,
    'ss': np.concatenate([env.init_qpos[1:],env.init_qvel]),
}

perturbation_values = [0, .1, 10]
lqr_gym(hopper_config, perturb=perturbation_values)
perturbation_values = [100, 1000, 10000]
lqr_gym(hopper_config, perturb=perturbation_values)
```

```python
def sim_x(x, u):
    x1_ = -2 * x[0] + a * x[1] + np.sin(x[0])
    x2_ = -x[1] * np.cos(x[0]) + u * np.cos(2 * x[0])
    return np.array([x1_, x2_])
```

```python
def x_to_z(x):
    return np.array([x[0], a * x[1] + np.sin(x[0])])
```

```python
def sim_z(z, v):
    z1_ = -2 * z[0] + z[1]
    z2_ = v
    return np.array([z1_, z2_])
```

```python
def v_to_u(v, z):
    return (v - np.cos(z[0]) * np.sin(z[0]) + 2 * z[0] * np.cos(z[0])) / (a * np.cos(2 * z[0]))
```

```python
class iLQR_Solver:
    def __init__(self, T, f, cost_fn, aux=None):
        self.T = T
        self.f = f
        self.cost_fn = cost_fn   
        self.env = aux
        self.aux=None
        self.plant_dyn_dx = None        #Df/Dx
        self.plant_dyn_du = None        #Df/Du
        self.cost_dx = None             #Dl/Dx
        self.cost_du = None             #Dl/Du
        self.cost_dxx = None            #D2l/Dx2
        self.cost_duu = None            #D2l/Du2
        self.cost_dux = None            #D2l/DuDx

        self.constraints_dx = None      #Dc/Dx
        self.constraints_du = None      #Dc/Du
        self.constraints_dxx = None     #D2c/Dx2
        self.constraints_duu = None     #D2c/Du2
        self.constraints_dux = None     #D2c/DuDx

        self.finite_diff_eps = .0001
        #candidate alphas for line search in the direction of gradients
        #10 line steps
        self.reg=.1
        self.alpha_array = 1.1 ** (-np.arange(10)**2)   
        #adapting the regularizer
        self.reg_max = 1000
        self.reg_min = 1e-6
        self.reg_factor = 10
        return

    def evaluate_trajectory_cost(self, x_array, u_array):
        #Note x_array contains X_T, so a dummy u is required to make the arrays 
        #be of consistent length
        u_array_sup = np.vstack([u_array, np.zeros(len(u_array[0]))])
        J_array = [self.cost_fn(x, u, t) for t, (x, u) in enumerate(zip(x_array, u_array_sup))]
        return np.sum(J_array)

    def ilqr_iterate(self, x0, u_init, n_itrs=50, tol=1e-6, verbose=True):
        self.reg=1
        #derive the initial guess trajectory from the initial guess of u
        x_array = self.forward_propagation(x0, u_init)
        u_array = np.copy(u_init)
        #initialize current trajectory cost
        J_opt = self.evaluate_trajectory_cost(x_array, u_init)
        J_hist = [J_opt]
        #iterates...
        converged = False
        for i in range(n_itrs):
            k_array, K_array = self.back_propagation(x_array, u_array)
            norm_k = np.mean(np.linalg.norm(k_array, axis=1))
            #apply the control to update the trajectory by trying different alpha
            accept = False
            for alpha in self.alpha_array:
                x_array_new, u_array_new = self.apply_control(x_array, u_array, k_array, K_array, alpha)
                #evaluate the cost of this trial
                J_new = self.evaluate_trajectory_cost(x_array_new, u_array_new)

                if J_new < J_opt:
                    #see if it is converged
                    if np.abs((J_opt - J_new )/J_opt) < tol:
                        #replacement for the next iteration
                        J_opt = J_new
                        x_array = x_array_new
                        u_array = u_array_new
                        converged = True
                        break
                    else:
                        #replacement for the next iteration
                        J_opt = J_new
                        x_array = x_array_new
                        u_array = u_array_new
                        #successful step, decrease the regularization term
                        #momentum like adaptive regularization
                        self.reg = np.max([self.reg_min, self.reg / self.reg_factor])
                        accept = True
#                         if verbose:
                        print('Iteration {0}:\tJ = {1};\tnorm_k = {2};\treg = {3}'.format(i+1, J_opt, norm_k, np.log10(self.reg)))
                        break
                else:
                    #don't accept this
#                     print('Converged at iteration {0}; J = {1}; reg = {2}'.format(i+1, J_opt, self.reg))
                    accept = False
            
            J_hist.append(J_opt)

            #exit if converged...
            if converged:
                if verbose:
                    print('Converged at iteration {0}; J = {1}; reg = {2}'.format(i+1, J_opt, self.reg))
                break
            #see if all the trials are rejected
            if not accept:
                #need to increase regularization
                #check if the regularization term is too large
                if self.reg > self.reg_max:
                    if verbose:
                        print('Exceeds regularization limit at iteration {0}; terminate the iterations'.format(i+1))
                    break

                self.reg = self.reg * self.reg_factor
                if verbose:
                    print('Reject the control perturbation. Increase the regularization term.')

        #prepare result dictionary
        res_dict = {
        'J_hist':np.array(J_hist),
        'x_array_opt':np.array(x_array),
        'u_array_opt':np.array(u_array),
        'k_array_opt':np.array(k_array),
        'K_array_opt':np.array(K_array)
        }

        return res_dict

    def apply_control(self, x_array, u_array, k_array, K_array, alpha):
        """
        apply the derived control to the error system to derive new x and u arrays
        """
        x_new_array = [None] * len(x_array)
        u_new_array = [None] * len(u_array)

        x_new_array[0] = x_array[0]
        for t in range(self.T):
#             u_new_array[t] = [0]
            u_new_array[t] = u_array[t] + alpha * (k_array[t] + K_array[t].dot(x_new_array[t] - x_array[t]))
#             print(u_new_array[t])
            x_new_array[t+1] = self.f(x_new_array[t], u_new_array[t], t)
            if t%10==0:
#                 print(x_new_array[t+1])
                plt.clf()
                plt.axis('off')
                plt.grid(b=None)
                plt.imshow(self.env.render(mode='rgb_array', width=64, height=64))
                plt.title("Alpha {}".format(alpha))
                display.clear_output(wait=True)
                display.display(plt.gcf())

        return np.array(x_new_array), np.array(u_new_array)

    def forward_propagation(self, x0, u_array):
        """
        Apply the forward dynamics to have a trajectory starting from x0 by applying u

        u_array is an array of control signal to apply
        """
        traj_array = [x0]

        for t, u in enumerate(u_array):
            traj_array.append(self.f(traj_array[-1], u, t))

        return traj_array

    def back_propagation(self, x_array, u_array):
        """
        Back propagation along the given state and control trajectories to solve
        the Riccati equations for the error system (delta_x, delta_u, t)
        Need to approximate the dynamics/costs/constraints along the given trajectory
        dynamics needs a time-varying first-order approximation
        costs and constraints need time-varying second-order approximation
        """
        #Note x_array contains X_T, so a dummy u is required to make the arrays 
        #be of consistent length
        u_array_sup = np.vstack([u_array, np.zeros(len(u_array[0]))])
        lqr_sys = self.build_lqr_system(x_array, u_array_sup)

        #k and K
        fdfwd = [None] * self.T
        fdbck_gain = [None] * self.T

        #initialize with the terminal cost parameters to prepare the backpropagation
        Vxx = lqr_sys['dldxx'][-1]
        Vx = lqr_sys['dldx'][-1]

        for t in reversed(range(self.T)):
            #note to double check if we need the transpose or not
            Qx = lqr_sys['dldx'][t] + lqr_sys['dfdx'][t].T.dot(Vx)
            Qu = lqr_sys['dldu'][t] + lqr_sys['dfdu'][t].T.dot(Vx)
            Qxx = lqr_sys['dldxx'][t] + lqr_sys['dfdx'][t].T.dot(Vxx).dot(lqr_sys['dfdx'][t])
            Qux = lqr_sys['dldux'][t] + lqr_sys['dfdu'][t].T.dot(Vxx).dot(lqr_sys['dfdx'][t])
            Quu = lqr_sys['dlduu'][t] + lqr_sys['dfdu'][t].T.dot(Vxx).dot(lqr_sys['dfdu'][t])

            #use regularized inverse for numerical stability
#             inv_Quu = np.linalg.pinv(Quu)
            inv_Quu = self.regularized_persudo_inverse_(Quu, reg=self.reg)
            #get k and K
            fdfwd[t] = -inv_Quu.dot(Qu)
            fdbck_gain[t] = -inv_Quu.dot(Qux)

            #update value function for the previous time step
            Vxx = Qxx - fdbck_gain[t].T.dot(Quu).dot(fdbck_gain[t])
            Vx = Qx - fdbck_gain[t].T.dot(Quu).dot(fdfwd[t])

        return fdfwd, fdbck_gain

    def build_lqr_system(self, x_array, u_array):
        dfdx_array = []
        dfdu_array = []
        dldx_array = []
        dldu_array = []
        dldxx_array = []
        dldux_array = []
        dlduu_array = []

        for t, (x, u) in enumerate(zip(x_array, u_array)):
            #refresh all the points for potential finite difference
            x1 = None
            x2 = None
            u1 = None
            u2 = None

            #for fx
            #use finite difference
            if x1 is None or x2 is None:
                x1 = np.tile(x, (len(x), 1)) + np.eye(len(x)) * self.finite_diff_eps
                x2 = np.tile(x, (len(x), 1)) - np.eye(len(x)) * self.finite_diff_eps
            fx1 = np.array([self.f(x1_dim, u, t, self.aux) for x1_dim in x1])
            fx2 = np.array([self.f(x2_dim, u, t, self.aux) for x2_dim in x2])
            dfdx_array.append((fx1-fx2).T/2./self.finite_diff_eps)

            #for fu
            #use finite difference
            if u1 is None or u2 is None:
                u1 = np.tile(u, (len(u), 1)) + np.eye(len(u)) * self.finite_diff_eps
                u2 = np.tile(u, (len(u), 1)) - np.eye(len(u)) * self.finite_diff_eps
            fu1 = np.array([self.f(x, u1_dim, t, self.aux) for u1_dim in u1])
            fu2 = np.array([self.f(x, u2_dim, t, self.aux) for u2_dim in u2])
            dfdu_array.append((fu1-fu2).T/2./self.finite_diff_eps)

            #for lx
            #use finite difference
            if x1 is None or x2 is None:
                x1 = np.tile(x, (len(x), 1)) + np.eye(len(x)) * self.finite_diff_eps
                x2 = np.tile(x, (len(x), 1)) - np.eye(len(x)) * self.finite_diff_eps
            cx1 = np.array([self.cost_fn(x1_dim, u, t, self.aux) for x1_dim in x1])
            cx2 = np.array([self.cost_fn(x2_dim, u, t, self.aux) for x2_dim in x2])
            dldx_array.append((cx1-cx2).T/2./self.finite_diff_eps)

            #for lu
            #use finite difference
            if u1 is None or u2 is None:
                u1 = np.tile(u, (len(u), 1)) + np.eye(len(u)) * self.finite_diff_eps
                u2 = np.tile(u, (len(u), 1)) - np.eye(len(u)) * self.finite_diff_eps
            cu1 = np.array([self.cost_fn(x, u1_dim, t, self.aux) for u1_dim in u1])
            cu2 = np.array([self.cost_fn(x, u2_dim, t, self.aux) for u2_dim in u2])
            dldu_array.append((cu1-cu2).T/2./self.finite_diff_eps)

            #for lxx
            #use finite difference
            # l = self.cost(x, u, t, self.aux)
            # dldxx_array.append(np.array([[(cx1_dim + cx2_dim - 2*l)/(self.finite_diff_eps**2) for cx2_dim in cx2] for cx1_dim in cx1]))
            dldxx_array.append(
                self.finite_difference_second_order_(
                    lambda x_arg: self.cost_fn(x_arg, u, t, self.aux),
                    x))

            #for luu
            #use finite difference
            # l = self.cost(x, u, t, self.aux)
            # dlduu_array.append(np.array([[(cu1_dim + cu2_dim - 2*l)/(self.finite_diff_eps**2) for cu2_dim in cu2] for cu1_dim in cu1])) 
            dlduu_array.append(
                self.finite_difference_second_order_(
                    lambda u_arg: self.cost_fn(x, u_arg, t, self.aux),
                    u))
            #for lux
            #use finite difference
            l = self.cost_fn(x, u, t, self.aux)
            cux1 = np.array([[self.cost_fn(x1_dim, u1_dim, t, self.aux) for x1_dim in x1] for u1_dim in u1])
            cux2 = np.array([[self.cost_fn(x2_dim, u2_dim, t, self.aux) for x2_dim in x2] for u2_dim in u2])
            #partial derivative - a simplified approximation, see wiki on finite difference
            dldux = cux1 + cux2 + \
                    2 * np.tile(l, (len(x), len(u))).T - \
                    np.tile(cx1, (len(u), 1)) - np.tile(cx2, (len(u), 1)) - \
                    np.tile(cu1, (len(x), 1)).T - np.tile(cu2, (len(x), 1)).T

            dldux_array.append(dldux/(2*self.finite_diff_eps**2)) 
            # print dfdx_array[-1], dfdu_array[-1], dldx_array[-1], dldu_array[-1]
            # print dldxx_array[-1], dlduu_array[-1], dldux_array[-1]
            # raw_input()
            
            #need to do somthing similar for constraints if they were there
            #to incorporate with the cost functions. Ignore them for now
        
        lqr_sys = {
            'dfdx':dfdx_array,
            'dfdu':dfdu_array,
            'dldx':dldx_array,
            'dldu':dldu_array,
            'dldxx':dldxx_array,
            'dlduu':dlduu_array,
            'dldux':dldux_array
            }

        return lqr_sys
    
    def regularized_persudo_inverse_(self, mat, reg=1e-5):
        """
        Use SVD to realize persudo inverse by perturbing the singularity values
        to ensure its positive-definite properties
        """
        u, s, v = np.linalg.svd(mat)
        s[ s < 0 ] = 0.0        #truncate negative values...
        diag_s_inv = np.zeros((v.shape[0], u.shape[1]))
        diag_s_inv[0:len(s), 0:len(s)] = np.diag(1./(s+reg))
        return v.dot(diag_s_inv).dot(u.T)
    
    def finite_difference_second_order_(self, func, x):
        n_dim = len(x)
        func_x = func(x)
        hessian = np.zeros((n_dim, n_dim))
        for i in range(n_dim):
            for j in range(n_dim):
                x_copy = np.copy(x)
                x_copy[i] += self.finite_diff_eps
                x_copy[j] += self.finite_diff_eps
                fpp = func(x_copy)

                x_copy = np.copy(x)
                x_copy[i] += self.finite_diff_eps
                fp_ = func(x_copy)

                x_copy = np.copy(x)
                x_copy[j] += self.finite_diff_eps
                f_p = func(x_copy)

                x_copy = np.copy(x)
                x_copy[i] -= self.finite_diff_eps
                fn_ = func(x_copy)

                x_copy = np.copy(x)
                x_copy[j] -= self.finite_diff_eps
                f_n = func(x_copy)
        
                x_copy = np.copy(x)
                x_copy[i] -= self.finite_diff_eps
                x_copy[j] -= self.finite_diff_eps
                fnn = func(x_copy)

                hessian[i, j] = fpp - fp_ - f_p - f_n - fn_ + fnn

        hessian = (hessian + 2*func_x) / (2*self.finite_diff_eps**2)
        return hessian
```

```python
class PendulumProblem():
    def __init__(self, T=150):
        env = InvertedPendulumEnv()
#         env = Walker2dEnv()
#         env = HopperEnv()
#         env = HalfCheetahEnv()
        self.x0=np.concatenate([env.init_qpos[:],env.init_qvel])
        self.ilqr = None
        self.res = None
        self.T = T

        self.Q = 100
        self.R = .01

        #terminal Q to regularize final speed
        self.Q_T = .1
        return

    def f(self, x, u, t, aux=None):
#         nq, nv = env.model.nq, env.model.nv
#         env.sim.reset()
#         qpos = copy.deepcopy(env.init_qpos)
#         qvel = copy.deepcopy(env.init_qvel)

#         qpos[1:] = x[:nq-1]
#         qvel[:] = x[nq-1:]

#         env.set_state(qpos, qvel)
# #         print(u)
#         env.step(u)
#         return np.concatenate([
#             env.sim.data.qpos.flat[1:],
#             env.sim.data.qvel.flat[:]
#         ])
        nq, nv = env.model.nq, env.model.nv
        env.sim.reset()
        qpos = copy.deepcopy(env.init_qpos)
        qvel = copy.deepcopy(env.init_qvel)
#         print(qvel)
        qpos[:] = x[:nq]
        qvel[:] = x[nq:]
#         qpos[1:6] = x[:nq-1]
#         qvel[:6] = x[nq-1:]

        env.set_state(qpos, qvel)
        env.step(u)
        return np.concatenate([
            env.sim.data.qpos.flat[:],
            env.sim.data.qvel.flat[:]
        ])

    def instaneous_cost(self, x, u, t, aux=None):
#         posbefore = x[0]
#         self.do_simulation(a, self.frame_skip)
     #   posafter, height, ang = self.sim.data.qpos[0:3]
      #  alive_bonus = 1.0
#         reward = (x[0] - env.init_qpos[0]) / env.dt
       # reward += alive_bonus
#         reward -= 1e-3 * np.square(u).sum()
#         return -r/eward
#         s = x[1:]
#         done = not (np.isfinite(s).all() and (np.abs(s[2:]) < 100).all() and
#                     (x[1] > .7) and (abs(x[2]) < .2))
#         if not done:
#             reward += 1
#         return -reward
        return self.Q*(x[1] - self.x0[1])**2+np.linalg.norm(u)**2
#         return self.Q*np.linalg.norm(x - self.x0)+np.linalg.norm(u)

    #grad_types = ['user', 'autograd', 'fd']
    def build_ilqr_problem(self, grad_type=0):
        self.ilqr = iLQR_Solver(T=self.T, f=self.f, cost_fn=self.instaneous_cost, aux = env)
        return

    def solve_ilqr_problem(self, x0=None, u_init=None, n_itrs=150, verbose=True):
        #prepare initial guess
        if u_init is None:
#             u_init = np.zeros((self.T, 6))
            u_init = np.random.normal(scale=0, size=(self.T, 1))
        if x0 is None:
            x0 = np.array([np.random.rand()*2*np.pi - np.pi, 0])

        if self.ilqr is not None:
            self.res = self.ilqr.ilqr_iterate(x0, u_init, n_itrs=n_itrs, tol=1e-6, verbose=verbose)
        return

    def plot_ilqr_result(self, name):
        if self.res is not None:
            #draw cost evolution and phase chart
            fig = plt.figure(figsize=(16, 8), dpi=80)
            ax_cost = fig.add_subplot(121)
            n_itrs = len(self.res['J_hist'])
            ax_cost.plot(np.arange(n_itrs), self.res['J_hist'], 'r', linewidth=3.5)
            ax_cost.set_xlabel('Number of Iterations', fontsize=20)
            ax_cost.set_ylabel('Trajectory Cost')

            ax_phase = fig.add_subplot(122)
            theta = self.res['x_array_opt'][:, 0]
            theta_dot = self.res['x_array_opt'][:, 1]
            ax_phase.plot(theta, theta_dot, 'k', linewidth=3.5)
            ax_phase.set_xlabel('theta (rad)', fontsize=20)
            ax_phase.set_ylabel('theta_dot (rad/s)', fontsize=20)
            ax_phase.set_title('Phase Plot', fontsize=20)

            ax_phase.plot([theta[-1]], [theta_dot[-1]], 'b*', markersize=16)

            plt.savefig(name)

        return
```
